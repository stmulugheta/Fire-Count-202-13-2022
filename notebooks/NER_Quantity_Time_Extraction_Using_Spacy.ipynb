```python
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
import en_core_web_lg
nlp= en_core_web_lg.load()
```

```python
data_df = pd.read_csv("clean_sentences.csv")
data_df
```

```python
list1=['PERCENT','QUANTITY']
sentence = [_ for _ in data_df['Sentences']]

quant = []
time=[]
for doc in nlp.pipe(sentence):
  quantity = []
  date1=[]
  for ent in doc.ents:
    if ent.label_ in list1:
      quantity.append(ent)
    if ent.label_ == "DATE":
      date1.append(ent)
  quant.append(quantity)
  time.append(date1)
```

```python
data_df['Quantity'] = pd.Series(quant)
data_df['Time'] = pd.Series(time)
```
