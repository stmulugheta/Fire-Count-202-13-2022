```python
import pandas as pd
df = pd.read_csv('KPI sample dataset.csv')
df
```

```python
from transformers import pipeline
classifier = pipeline(model='nbroad/ESG-BERT')
```

```python
result = classifier(df["Sentences"].to_list())
```

```python
sub_topic = [res['label'] for res in result]
df["ESG_BERT_subTopics"] = sub_topic

pd.set_option('display.max_rows', None)
pd.set_option('max_colwidth', 100)
df[["Sentences", "ESG_BERT_subTopics"]]
```
