Lets see how many companies and sentences are there from all _sent.csv files

```python
import os
files = [f for f in os.listdir('.') if os.path.isfile(f)]
dataframe = pd.DataFrame()
for f in files:
  if f.endswith('_sent.csv'):
    dataframe = dataframe.append(pd.read_csv(f))
```

```python
dataframe.company.value_counts()
```

```python
len(dataframe.company.unique())
```

```python
len(dataframe)
```

```python
dataframe.company.unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
company_count  = dataframe['company'].value_counts()
company_count = company_count[:30,]
plt.figure(figsize=(10,5))
plot = sns.barplot(company_count.index, company_count.values, alpha=0.8)
plt.title('Top 30 companies with most sentences')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('company', fontsize=12)
plt.setp(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
company_count  = dataframe['company'].value_counts().sort_values()
company_count = company_count[:20,]
plt.figure(figsize=(10,5))
plot = sns.barplot(company_count.index, company_count.values, alpha=0.8)
plt.title('companies with less sentences/data')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('company', fontsize=12)
plt.setp(plot.get_xticklabels(), rotation=90)
plt.show()
```

- There are totally 312900 records and 502 unique companies
