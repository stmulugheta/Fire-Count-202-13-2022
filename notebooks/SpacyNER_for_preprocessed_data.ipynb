```python
import pandas as pd
import numpy as np
import en_core_web_lg
nlp= en_core_web_lg.load()
```

```python
data_df = pd.read_csv("Final_Extracted_text_AllPages_5f4212d7-7bb7-45e5-9d74-3c1e26770e55_Final - Final_Extracted_text_AllPages_5f4212d7-7bb7-45e5-9d74-3c1e26770e55_Final.csv")
data_df
```

```python
sentences = list(data_df["Sent_extracted"])
print(data_df.shape)
data_df = data_df[data_df['Sent_extracted'].str.split().str.len().gt(7)]
print(data_df.shape)
```

```python
list1=['PERCENT','QUANTITY']
sentence = [_ for _ in data_df['Sent_extracted']]

quant = []
time=[]
for doc in nlp.pipe(sentence):
  quantity = []
  date1=[]
  for ent in doc.ents:
    if ent.label_ in list1:
      quantity.append(ent)
    if ent.label_ == "DATE":
      date1.append(ent)
  quant.append(quantity)
  time.append(date1)
```

```python
data_df = data_df.assign(Quantity=quant)
data_df = data_df.assign(Time=time)
```

```python
data_df
```

```python
data_df[data_df['Quantity'].str.len()>0]
data_df.Quantity = data_df.Quantity.apply(lambda y: np.nan if len(y)==0 else y)
data_df.Time = data_df.Time.apply(lambda y: np.nan if len(y)==0 else y)
```
